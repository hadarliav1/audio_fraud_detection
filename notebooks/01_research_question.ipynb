# Stage 1 — Research Question

## AI-Generated Voice Fraud Detection: A Research-Grade Investigation

This notebook defines the research question, business motivation, and scientific rationale for the project.

## 1. Business Motivation

**AI-generated voice fraud** (voice cloning, deepfake audio) poses a growing threat across multiple domains:

- **Vishing (voice phishing):** Attackers use cloned voices of executives or family members to extract sensitive information or authorise fraudulent transfers.
- **Impersonation:** Political figures, celebrities, or business leaders can be falsely depicted saying things they never said.
- **Regulatory and compliance:** Financial services, healthcare, and government sectors need reliable methods to verify that voice interactions are with real humans, not synthetic impostors.

Modern TTS and voice conversion systems (e.g. ElevenLabs, VALL-E, voice cloning APIs) produce increasingly natural-sounding speech, making detection a moving target. Robust, scientifically justified detection methods are essential for deploying trustworthy voice authentication and fraud prevention systems.

## 2. Why Robustness Under Noise Matters

Real-world voice channels introduce **additive noise** (background sounds, ambient room noise) and **distortions** (telephony compression, packet loss, re-sampling). Detection systems that perform well only on clean studio recordings are of limited practical value.

We must evaluate:
- **White noise:** Worst-case baseline; flat spectrum stresses frequency-generalisation.
- **Pink noise:** 1/f spectrum; more speech-like; tests robustness to realistic environmental noise.
- **Compression artifacts:** MP3/AAC re-encoding mimics telephony and VoIP channels; tests whether models rely on high-frequency cues that are lost in real deployment.

Robustness at varying **SNR levels** (e.g. 20, 10, 5, 0 dB) quantifies degradation and guides deployment decisions.

## 3. Why Feature Fusion Is Scientifically Interesting

**Handcrafted acoustic features** (MFCCs, jitter, HNR, spectral flux, etc.) encode low-level signal properties that may reveal vocoder artifacts or unnatural regularity in synthetic speech. **Transformer embeddings** capture high-level semantic and prosodic patterns from self-supervised pre-training.

These representations are **partially orthogonal**: acoustic features emphasise physical signal characteristics; transformers emphasise learned linguistic/prosodic structure. Combining them may:
- Capture complementary cues (e.g. vocoder artifacts + contextual inconsistency)
- Improve robustness if acoustic features degrade more gracefully under noise than transformers
- Provide interpretability (feature importance) alongside transformer performance

We will **rigorously test** whether fusion adds value beyond either representation alone—no assumption of superiority.

## 4. Main Research Question

> **Can we detect AI-generated speech with sufficient accuracy and robustness under real-world noise conditions, and does combining handcrafted acoustic features with transformer embeddings improve detection?**

### Sub-Questions

1. **Which acoustic cues best discriminate real vs. synthetic speech?** — EDA and feature importance analysis.

2. **Do transformer models outperform classical/CNN baselines?** — Fair comparison with consistent evaluation metrics.

3. **Does feature fusion add value beyond transformers alone?** — Ablation: transformer-only vs. acoustic-only vs. fused.

4. **How does performance degrade under white, pink, and compression noise at varying SNR?** — Degradation curves for all models.

## 5. Cost-Sensitive Framing

In fraud detection, the cost of **false negatives** (missed fraud) often exceeds the cost of **false positives** (blocked legitimate calls). We will:
- Monitor precision, recall, and F1 in addition to accuracy
- Consider class imbalance and adjust metrics (e.g. macro F1, cost-weighted loss) if needed
- Report confusion matrices to support threshold tuning for deployment
